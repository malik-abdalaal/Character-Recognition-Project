# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import layers
```

# Intro to the dataset

# Data

#### Load the dataset form files

```python
x_train = np.load('training_images.npy')
y_train = np.load('training_labels.npy')
x_test = np.load('test_images.npy')
y_test = np.load('test_labels.npy')
```

#### Normalization

```python
x_train = np.expand_dims(x_train, axis=-1)
x_train = x_train.astype("float32") / 255
x_test = np.expand_dims(x_test, axis=-1)
x_test = x_test.astype("float32") / 255
```

#### Sample of the dataset

```python
# Define the layout of subplots
num_rows = 5  # Number of rows of subplots
num_cols = 5  # Number of columns of subplots
num_images = num_rows * num_cols

# Create the figure and subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))
axes = axes.ravel()  # Flatten the axes object

# Number of wrong predictions
number_of_wrong_predictions = 0

# Display the images on subplots
for i in range(num_images):
    temp = x_train[i].reshape(32, 32)  # Reshape the image from flattened to 32x32
    temp = temp.reshape(32, 32)  # Reshape the image from flattened to 32x32
    label = chr(y_train[i])

    ax = axes[i]  # Get the appropriate subplot for the current image

    ax.imshow(temp, cmap='gray')
    ax.set_title(label)
    ax.axis('off')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#### Shift data

```python
y_train -= 65
y_test -= 65
```

#### Build the model

```python
model = keras.models.Sequential(name="chars_classifier")

# Input layer
model.add(layers.Input(shape=(32, 32, 1), name="chars_image"))

# Conv layers
model.add(layers.Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Flatten())

# Classi layers
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(26, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train, epochs=55, verbose=1, batch_size=128)
```

```python
from tabnanny import verbose
```

```python
print(f"Accuracy: {round(model.evaluate(x_test, y_test, verbose=False)[1], 2)}")
```

#### Save model weights

```python
model.save('model\\chars_model.h5')
```

#### Implement function to predict new characters

```python
def predict_image(image):
    if np.sum(image) == 0:
        return ' '
    
    # Load the saved model
    model = keras.models.load_model('model\\chars_model.h5')

    # Preprocess the image
    image = keras.preprocessing.image.img_to_array(image)
    image = image.reshape(1, 32, 32, 1).astype('float32') / 255.0

    # Make predictions
    predictions = model.predict(image, verbose=False)
    predicted_label = np.argmax(predictions[0])

    return chr(int(predicted_label + 65))
```

```python
def test_letters(font):
    # Define the layout of subplots
    num_rows = 2  # Number of rows of subplots
    num_cols = 13  # Number of columns of subplots
    num_images = num_rows * num_cols

    # Create the figure and subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 5))
    axes = axes.ravel()  # Flatten the axes object

    # Number of wrong predictions
    number_of_wrong_predictions = 0

    # Display the images on subplots
    for i in range(num_images):
        temp = image.imread(f'images\\{font}\\{chr(i + 65)}.jpg')[:,:, 0]
        temp = temp.reshape(32, 32)  # Reshape the image from flattened to 32x32
        label = predict_image(temp)
        
        if label != chr(i + 65):
            number_of_wrong_predictions += 1

        ax = axes[i]  # Get the appropriate subplot for the current image

        ax.imshow(temp, cmap='gray_r')
        ax.set_title(label)
        ax.axis('off')

    # Adjust the spacing between subplots
    plt.tight_layout()

    # Show the plot
    plt.show()

    # Print number of wrong predictions
    print(f'Number of wrong predictions: {number_of_wrong_predictions}')
```

```python
import os

directory = 'images'

for filename in os.listdir(directory):
    if os.path.join(directory, filename) != 'images\\text':
        test_letters(filename)
```

# Predicting full text

```python
def predict_text(test_image):
    text = ''

    height, width = test_image.shape
    num_of_chars = int(width / 32)

    for item in range(num_of_chars):
        text += predict_image(test_image[:, item * 32: item * 32 + 32])
    
    return text
```

```python
from PIL import Image, ImageDraw, ImageFont

def generate_text_image(text, font):
    # Image path
    image_path = f'images\\text\\{font[0:font.index(".")]}-{text}.jpg'

    # Create a blank image with black background
    image = Image.new("RGB", (32 * len(text), 32), "black")
    
    # Define the font and size
    font = ImageFont.truetype(f"{font}", 28, encoding='unic')

    for i in range(0, len(text)):
        letter = text[i]

        # Get the size of the letter
        letter_width, letter_height = font.getsize(letter)

        # Calculate the center coordinates of the image
        center_x = ((32 * (i + 1)) - 16)
        center_y = image.height // 2

        # Calculate the top-left coordinates of the letter
        letter_x = center_x - letter_width // 2
        letter_y = (center_y - letter_height // 2) - 2
        
        # Create a draw object
        draw = ImageDraw.Draw(image)

        # Draw the letter on the image
        draw.text((letter_x, letter_y), letter, font=font, fill="white")

    # Save the image as JPG
    image.save(image_path, "JPEG")
    
    return image_path
```

```python
def differences(str1, str2):
    indexes = []
    
    for i in range(0, len(str1)):
        if str1[i] != str2[i]:
            indexes.append(i)
            
    return indexes
```

```python
fonts = ['arialbd.ttf', 'bahnschrift.ttf', 'Candarab.ttf', 'Candaral.ttf', 'ebrimabd.ttf', 'SitkaVF.ttf', 'COLONNA.TTF', 'Gabriola.ttf', 'verdanab.ttf']
# fonts = [] # , 'FORTE.TTF'
```

```python
images = {}
    
for text in ['ABCDEFGHIJKLMNOPQRSTUVWXYZ']:
    for font in fonts:
        path = generate_text_image(text, font)
        images[path] = {'text': text, 'font': font}
```

#### Test the model on full text

```python
number_of_wrong_letters = 0
total_fonts = 0

for key in images.keys():
    total_fonts += 1
    
    temp = image.imread(key)[:,:,0]

    plt.figure(figsize=(15,5))
    plt.imshow(temp, cmap='gray_r')
    plt.show()

    prediction = predict_text(temp)

    print(images[key]['font'])
    print(f'predicted text: {prediction}')
    
    indexes = differences(prediction, images[key]['text'])
        
    if len(indexes) == 0:
        print('The predicted text is 100% true')
    else:
        number_of_wrong_letters += len(indexes)
        wrong = ' ' * 16
            
        for i in range(0, len(indexes)):
            if i > 0:
                wrong += ' ' * (indexes[i] - indexes[i - 1] - 1)
            else:
                wrong += ' ' * (indexes[i])
                
            wrong += '^'
        
        print(wrong)
        
print('==================================================================\n')
total_tested_characters = total_fonts * 26
print(f'Total letters: {total_tested_characters}')
print(f'Number of wrong letters: {number_of_wrong_letters}')
print(f'Acc: {round(((total_tested_characters - number_of_wrong_letters) / (total_tested_characters)) * 100, 2)}%')
```
